```python
ext = '.pdf'
path = 'images/histograms/'

def dobar(var,xlabel,title):
    var.plot.bar(alpha=0.75)
    plt.xlabel(xlabel)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.savefig(path+title+ext)
    plt.show()
    
def dohist(var,xlabel,title,bins=10):
    var.plot.hist(alpha=0.75,bins=bins)
    plt.xlabel(xlabel)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.savefig(path+title+ext)
    plt.show()
```
